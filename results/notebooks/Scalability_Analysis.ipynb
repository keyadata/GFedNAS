```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rcParams

rcParams.update(
    {
        "font.family": "sans-serif",
        "font.sans-serif": ["DejaVu Sans"],
        "axes.linewidth": 0.7,
        "axes.labelsize": 9,
        "axes.titlesize": 10,
        "xtick.labelsize": 8,
        "ytick.labelsize": 8,
        "grid.alpha": 0.35,
        "grid.linestyle": "--",
        "figure.dpi": 300,
    }
)

num_clients = np.array([10, 15, 20, 25, 30])
scal_acc    = np.array([78.56, 83.33, 89.01, 92.7, 94.0]) / 100.0
scal_energy = np.array([0.454, 0.499, 0.5037, 0.601, 0.774])  # Wh

fig, ax1 = plt.subplots(figsize=(5.4, 3.2))

# Accuracy â€” BLUE
ax1.plot(
    num_clients, scal_acc,
    marker='o', linewidth=2,
    label="Accuracy",
    color="#1f77b4"
)
ax1.set_xlabel("Number of Edge Devices (Clients)")
ax1.set_ylabel("Accuracy (%)")
ax1.set_ylim(0.50, 1.00)
ax1.set_yticks(np.arange(0.50, 1.01, 0.10))


ax1.set_xticks(num_clients)

ax1.grid(True)


ax2 = ax1.twinx()
ax2.plot(
    num_clients, scal_energy,
    marker='s', linestyle='--', linewidth=2,
    label="Energy (Wh)",
    color="#2ca02c"
)
ax2.set_ylabel("Energy Consumption (Wh)")


ax1.spines["top"].set_visible(False)
ax2.spines["top"].set_visible(False)


# Combined
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(
    h1 + h2, l1 + l2,
    fontsize=8, frameon=True,
    loc="lower right"
)

#ax1.set_title("Scalability Analysis")
plt.tight_layout()
plt.savefig("scalability.png", dpi=300, bbox_inches="tight")
plt.show()
```
